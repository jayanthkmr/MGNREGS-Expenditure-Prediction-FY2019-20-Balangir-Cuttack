```python
dist_all_data_no_geo = pd.merge(
    dist_expnd_df,
    odisha_dist_df_concat,
    how="inner",
    left_on="District",
    right_on="Name",
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_a", "_b"),
    copy=True,
    indicator=False,
    validate=None,
)
```

```python
dist_all_data = pd.merge(
    dist_all_data_no_geo,
    dist_area_df,
    how="inner",
    left_on="District_b",
    right_on="censuscode",
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_p", "_q"),
    copy=True,
    indicator=False,
    validate=None,
)
```

```python
dist_all_data = dist_all_data[sorted(dist_all_data)].sort_values(by=['District_a'])
```

```python
dist_all_data['PERCENT_EXP_TOT'] = dist_all_data['TOT_EXP_TOT']/(dist_all_data['TOT_EXP_TOT'].sum())
```

```python
dist_all_data['SCALED_EXP_TOT'] = (dist_all_data['PERCENT_EXP_TOT'] - dist_all_data['PERCENT_EXP_TOT'].mean()) / (dist_all_data['PERCENT_EXP_TOT'].std())
dist_all_data['MINMAX_EXP_TOT'] = (dist_all_data['PERCENT_EXP_TOT'] - dist_all_data['PERCENT_EXP_TOT'].min()) / (dist_all_data['PERCENT_EXP_TOT'].max() - dist_all_data['PERCENT_EXP_TOT'].min())
```

```python
# X = dist_all_data[dist_all_data.columns.difference(['UNSKL_EXP_20', 'UNSKL_EXP_PREV', 'UNSKL_EXP_TOT',
#        'MAT_EXP_20', 'MAT_EXP_PREV', 'MAT_EXP_TOT', 'ADMN_EXP_20',
#        'ADMN_EXP_PREV', 'ADMN_EXP_TOT', 'TOT_EXP_20', 'TOT_EXP_PREV', 'TOT_EXP_TOT', 'Area', 'Perimeter', 'Longitude','Latitude'])]
X = dist_all_data.select_dtypes(['number'])[['Area', 'Perimeter', 'Longitude','Latitude']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor

class LogitRegression(GradientBoostingRegressor):

    def fit(self, x, p):
        p = np.asarray(p)
        y = np.log(p / (1 - p)) 
        return super().fit(x, y)

    def predict(self, x):
        y = super().predict(x)
        return 1 / (np.exp(-y) + 1)
```

```python
from sklearn.multioutput import MultiOutputRegressor, RegressorChain
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler, Normalizer, MinMaxScaler
regr = make_pipeline(MinMaxScaler(), LogitRegression())
#base_regr = RadiusNeighborsRegressor(radius=1.0, weights="uniform")
#regr = make_pipeline(StandardScaler(), MultiOutputRegressor(base_regr))
# regr = make_pipeline(MinMaxScaler(), RegressorChain(base_estimator=base_regr, 
#                                                       order= [0, 1, 8, 2, 3, 9, 4, 5, 10, 6, 7, 11]))
regr.fit(X, Y)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
# define the evaluation procedure
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate the model and collect the scores
n_scores = cross_val_score(regr, X, Y, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1)
# force the scores to be positive
n_scores = abs(n_scores)
# summarize performance
print('MAE: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```

```python
Y_pred= regr.predict(X)
```

```python
from sklearn.metrics import explained_variance_score
explained_variance_score(Y, Y_pred, multioutput='raw_values')
```
